```python
import seaborn as sns
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 

from seaborn import relplot

```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.relplot(x = 'total_bill', y = 'tip', data = tips)
```

```python
sns.relplot(x  = tips['total_bill'], y = tips['tip'], data = tips, hue = 'smoker')
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="smoker", style="time")
```

```python
tips['size'].value_counts()
```

```python
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", size="size", sizes = (20, 100))
```

```python
from numpy.random import randn
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(dict(time = np.arange(500), value = randn(500).cumsum()))
df
```

```python
relplot(x = 'time', y = 'value', kind = 'line', data = df)
```

```python
fmri = sns.load_dataset('fmri')
fmri.head()
```

```python
# fig, ax = plt.subplots()
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri)
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, label = 'signal', height = 4, aspect=2)
plt.legend(loc = 'right')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line',estimator = None,  data = fmri, label = 'signal', height = 4, aspect=2)
plt.legend(loc = 'right')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue='event', label = 'signal', height = 4, aspect=2)
plt.legend(loc = 'right')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'line', data = fmri, hue='event',  style = 'region', label = 'signal', height = 4, aspect=2)
plt.legend(loc = 'right')
```

```python
sns.relplot(x = 'timepoint', y = 'signal', kind = 'scatter', data = fmri, hue='event',  style = 'region', label = 'signal', height = 4, aspect=2)
plt.legend(loc = 'right')
```
