```python
import pandas as pd
df = pd.read_csv('Data/nba.csv')
```

```python
df['Name'] = df['Name'].astype(str)
df['Name'].apply(len)
```

```python



def make_double(x):
    if type(x) == str:
        x = x + " " + x
        
    if type(x) == int:
        x = x * 2
    
    return x


df['Name'].apply(lambda x : make_double(x))
```

```python
def combine(x, y):
    return x + " " + str(y)


df.apply(lambda x: combine(x['Name'], x['Team']), axis = 1)
```
