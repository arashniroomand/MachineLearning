# Matplotlib Plot Structure  

- **Minor tick**: Smaller ticks between major ticks for finer detail.  
- **Major tick label**: Displays the scale values on the axes.  
- **Marker**: Symbol used in scatter plots to represent data points.  
- **Legend**: Box (often at the top) explaining plot elements.  
- **Title**: Heading describing the plot's purpose.

````python
import matplotlib.pyplot as plt

import numpy as np 
import pandas as pd 

from numpy.random import randint 

x = np.linspace(1,10, 20)
y = randint(1,50, 20)
print(x, y)
x.size, y.size
````

````python
plt.plot(x,y)
````

````python
plt.plot(x, y)
````

The difference is in whether you explicitly create a new figure or use the default one:  

### `plt.plot(x, y)`  
- Uses the **current active figure** (or creates one automatically if none exists).  
- If you call `plt.plot()` multiple times, lines will be added to the same figure unless cleared.  

### `plt.figure()` + `plt.plot(x, y)`  
- `plt.figure()` **explicitly creates a new empty figure** before plotting.  
- Ensures a fresh figure (avoids overlapping with previous plots).  

### Which to use?  
- Use `plt.plot(x, y)` for quick plotting.  
- Use `plt.figure()` first if you want to **start a new plot** (e.g., in a loop or multiple separate plots).  

Example:  
```python
plt.figure()  # Creates a new blank figure  
plt.plot(x, y)  # Plots on that figure  
```  
Without `plt.figure()`, repeated `plt.plot()` calls may draw on the same figure.
