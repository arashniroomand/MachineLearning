# Matplotlib Plot Structure  

- **Minor tick**: Smaller ticks between major ticks for finer detail.  
- **Major tick label**: Displays the scale values on the axes.  
- **Marker**: Symbol used in scatter plots to represent data points.  
- **Legend**: Box (often at the top) explaining plot elements.  
- **Title**: Heading describing the plot's purpose.

````python
import matplotlib.pyplot as plt

import numpy as np 
import pandas as pd 

from numpy.random import randint 

x = np.linspace(1,10, 20)
y = randint(1,50, 20)
print(x, y)
x.size, y.size
````

````python
plt.plot(x,y)
````

````python
plt.plot(x, y)
````

The difference is in whether you explicitly create a new figure or use the default one:  

### `plt.plot(x, y)`  
- Uses the **current active figure** (or creates one automatically if none exists).  
- If you call `plt.plot()` multiple times, lines will be added to the same figure unless cleared.  

### `plt.figure()` + `plt.plot(x, y)`  
- `plt.figure()` **explicitly creates a new empty figure** before plotting.  
- Ensures a fresh figure (avoids overlapping with previous plots).  

### Which to use?  
- Use `plt.plot(x, y)` for quick plotting.  
- Use `plt.figure()` first if you want to **start a new plot** (e.g., in a loop or multiple separate plots).  

Example:  
```python
plt.figure()  # Creates a new blank figure  
plt.plot(x, y)  # Plots on that figure  
```  
Without `plt.figure()`, repeated `plt.plot()` calls may draw on the same figure.

````python
import pandas as pd 

imdb = pd.read_csv('data/IMDB-Movie-Data.csv').head(50).fillna(0)
imdb.isnull().sum()
````

````python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.subplot(2, 2, 1)
plt.plot(imdb['Rank'], imdb['Rating'], linestyle = '--', marker = 'o',)

plt.xlabel('Rank')
plt.ylabel('Rating')
plt.title('Rank vs Rating')


plt.subplot(2, 2, 2)
plt.plot(imdb['Rank'], imdb['Revenue (Millions)'], color = 'g', linestyle = '--',
         marker = 'o', markerfacecolor = 'y', markeredgecolor = 'b')

plt.xlabel('Rank')
plt.ylabel('Revenue')
plt.title('Rank vs Revenue')


plt.subplot(2, 2, 2)
plt.plot(imdb['Rank'], imdb['Revenue (Millions)'], color = 'g', linestyle = '--',
         marker = 'o', markerfacecolor = 'y', markeredgecolor = 'b')

plt.xlabel('Rank')
plt.ylabel('Revenue')
plt.title('Rank vs Revenue')


plt.subplot(2, 2, 3)
plt.plot(imdb['Rank'], imdb['Metascore'], color = 'r', linestyle = '--',
         marker = 'o', markerfacecolor = 'g', markeredgecolor = 'b')

plt.xlabel('Rank')
plt.ylabel('metascore')
plt.title('Rank vs Metascore')


plt.subplot(2, 2, 4)
plt.plot(imdb['Rank'], imdb['Votes'], color = 'r', linestyle = '--',
         marker = 'o', markerfacecolor = 'y', markeredgecolor = 'b')

plt.xlabel('Rank')
plt.ylabel('Votes')
plt.title('Rank vs Votes')

plt.tight_layout()
````

````python
import matplotlib.pyplot as plt

# Create figure and axes (note: subplots() instead of subplot())
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))

# First plot (Rank vs Rating)
ax[0].plot(imdb['Rank'], imdb['Rating'], 
          linestyle='--', 
          marker='o',
          color='blue')
ax[0].set_xlabel('Rank')
ax[0].set_ylabel('Rating')
ax[0].set_title('Rank vs Rating')
ax[0].grid(True, linestyle='--', alpha=0.7)

# Second plot (Rank vs Revenue)
ax[1].plot(imdb['Rank'], imdb['Revenue (Millions)'], 
          color='green', 
          linestyle='--',
          marker='o', 
          markerfacecolor='yellow', 
          markeredgecolor='blue')
ax[1].set_xlabel('Rank')
ax[1].set_ylabel('Revenue (Millions)')
ax[1].set_title('Rank vs Revenue')
ax[1].grid(True, linestyle='--', alpha=0.7)

# Adjust layout to prevent overlapping
plt.tight_layout()

# Show the plot
plt.show()
````
