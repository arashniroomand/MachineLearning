```python
x = iris[['sepal_width', 'sepal_length']]
y = iris[['petal_width', 'petal_length']]
```

```python
ax = y.plot()
x.plot(ax = ax)
```

```python

plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 150
ax = x.plot()
ax2 = y.plot(ax = ax, secondary_y=True)

ax.set_ylabel('sepal ( width and length)')
ax2.set_ylabel('petal ( width and length)')

ax.set_xlabel('#Samples')
ax.set_title('Secondary Axis Plot')


```

```python
import numpy as np 
import pandas as pd 
import seaborn as sns

import matplotlib.pyplot as plt 

```

```python
iris = sns.load_dataset('iris')
```

```python
df = iris.iloc[0][0:4]
```

```python
df.plot()
```

```python
df.plot(kind = 'bar')
```

```python
df.plot.bar(color = 'green', fontsize = 10, title = 'IRIS Bar Plot', ylabel = 'length/width', grid = True, ylim = (0, 10))
```

```python
df.plot.bar(color = 'green', fontsize = 10, title = 'IRIS Bar Plot', ylabel = 'length/width', grid = True, ylim = (0, 10), rot = 20)

plt.figure()
df.plot.barh(color = 'green', fontsize = 10, title = 'IRIS Bar Plot', ylabel = 'length/width', grid = True, ylim = (0, 10), rot = 20)
```

```python
df.plot.pie(title = 'IRIS Bar Plot', ylabel = 'length/width', rot = 20)
```

```python
iris.hist(color = 'b', grid = True, alpha = 0.6)

plt.tight_layout()
plt.show()
```
